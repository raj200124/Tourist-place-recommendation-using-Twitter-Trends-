```python
df["date"] = pd.to_datetime(df["date"]).dt.date
df = df.sort_values(by="date")
```

```python
def tweets_per_day(df):
    df['date'] = pd.to_datetime(df['date'])
    return df[['content']].groupby(df['date'].dt.date).count()
  # return df['Tweets'].groupby(df['Date'].dt.date).count() 
  # if you want output to be `Series`

df=tweets_per_day(df)
df
```

```python
df.plot(figsize=(20,10),grid=True)
pyplot.show()
```

```python

from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(df, model='multiplicative',period = 7)
fig, (ax1,ax2,ax3)=plt.subplots(3,1,figsize=(15,8))
result.trend.plot(ax=ax1)
result.resid.plot(ax=ax2)
result.seasonal.plot(ax=ax3)
```
